```python
def build_youtube_client(api_name: str, api_version: str, developer_key: str) -> googleapiclient.discovery.Resource:
    """
    Build and return a YouTube API client.

    Returns:
        A Google API Resource
    
    """
    try:
        logger.info("Building YouTube API client")
        return googleapiclient.discovery.build(api_name, api_version, developerKey=developer_key)
    except Exception as e:
        logger.exception("Failed to build YouTube API client")
        raise
```
